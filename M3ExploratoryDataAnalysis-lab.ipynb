## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
# Plotting box plot for the column 'ConvertedComp'
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['ConvertedComp'])
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
# Calculating Inter Quartile Range (IQR) for the column 'ConvertedComp'
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
print('Inter Quartile Range (IQR) for ConvertedComp:', IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
# Upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print('Lower Bound:', lower_bound)
print('Upper Bound:', upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
# Identifying the number of outliers
outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]
print('Number of Outliers in ConvertedComp:', outliers.shape[0])
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# Creating new DataFrame without outliers
df_no_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
```

```python
df_no_outliers["ConvertedComp"].mean()
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
# Correlation between 'Age' and all other numeric columns
correlation = df_no_outliers.corr()['Age'].sort_values(ascending=False)
print(correlation)
```
